**Spliting and Renaming the GEO_csv dataset** 

The values on the df1-table above are presented in one cell, therefore the cell needs to be splited and expanded. since the code returns column names as numbers(0,1,2,3....), columns have to be renamed to reflect the original column names

```python
# column named has been renamed to geo to shorten the column name
url1 = '/content/cells_geo.csv'
df1 = pd.read_csv(url1)
df1.columns = ['geo']
df1.head(8)
df6 = df1['geo'].str.split(";", n =11, expand = True )
#df6.drop(['0'], axis= 1, inplace= True)

df6.columns = ['index','CITY', 'STATUS', 'LOCALISATION', 'DECOUPZONE', 'ZONENAME', 'LONGITUDE', 'LATITUDE', 'REGION', 'AREA', 'CELL_ID', 'SITE_CODE']
df6.drop(['index'], axis= 1, inplace= True)
df6.head(5)
```

**Droping and Remaning columns per dataset**

```python
df2.drop(['DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B'], axis = 1, inplace = True)
df2.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'CELL_ID', 'SITE_CODE']
df2.head(10)
```

```python
df3.drop(['DW_A_NUMBER', 'DW_B_NUMBER', 'COUNTRY_A', 'COUNTRY_B',], axis = 1, inplace = True)
df3.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'CELL_ID', 'SITE_CODE']
df3.head(10)
```

```python
df4.drop(['DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B'], axis = 1, inplace = True)
#print(df4)
df4.head(10)
# Remaming columns correctly
df4.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'CELL_ID', 'SITE_CODE']
df4
```

**Merging Telcom_dataset tables for the three days**

```python
df7 = pd.concat([df2,df3, df4], axis = 0, ignore_index= True)
df7
```

```python
df6.merge(df7, how = 'left', on = 'SITE_CODE')
df7.drop_duplicates(subset= ['SITE_CODE'], inplace = True)
df8= df6.merge(df7, how = 'left', on = 'SITE_CODE')
df8.head(8)
```
